19. Gęstość powietrza atmosferycznego zmienia się w zależności od wysokości, malejąc wraz ze wzrostem wysokości. 
(a) Korzystając z danych podanych w tabeli, uzyskaj zależność dla zmiany gęstości w zależności od wysokości i oblicz gęstość na wysokości 7000 m. 
(b) Oblicz masę atmosfery, korzystając z uzyskanej korelacji. Załóż, że Ziemia jest doskonałą sferą o promieniu 6377 km i przyjmij, że grubość atmosfery wynosi 25 km.

(a) Korzystając z danych podanych w tabeli, uzyskaj zależność dla zmiany gęstości w zależności od wysokości i oblicz gęstość na wysokości 7000 m.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial
from scipy.integrate import quad


# Provided data
# z, km; ro, kg/m^3
data = [[6377, 6378, 6379, 6380, 6381, 6382, 6383, 6385, 6387, 6392, 6397, 6402],
        [1.225, 1.112, 1.007, 0.9093, 0.8194, 0.7364, 0.6601, 0.5258, 0.4135, 0.1948, 0.08891, 0.04008]]

z_km = np.array(data[0]) - data[0][0]
z = z_km * 1000
#z = np.array(data[0]) - data[0][0]

ro = np.array([1.225, 1.112, 1.007, 0.9093, 0.8194, 0.7364, 0.6601, 0.5258, 0.4135, 0.1948, 0.08891, 0.04008])

# Fit a 2nd order polynomial to the data
p = Polynomial.fit(z, ro, 2)

# Coefficients of the fitted polynomial
coefficients = p.convert().coef

# Plotting the data and the fitted polynomial
z_line = np.linspace(z.min(), z.max(), 100)
ro_fit = p(z_line)

plt.figure(figsize=(5, 3))
plt.plot(z, ro, 'bo', label='Data')
plt.plot(z_line, ro_fit, 'r-', label='Fitted 2nd order polynomial')
plt.xlabel('z (km)')
plt.ylabel('ro (kg/m^3)')
plt.title('Fitted 2nd Order Polynomial to Density Data')
plt.legend()
plt.grid(True)
plt.show()

# Coefficients for the fitted polynomial
a0, a1, a2 = coefficients

# LaTeX representation of the polynomial equation
equation_latex = r"$$\rho(z) = 2.237 \times 10^{-3} z^2 + (-1.017 \times 10^{-1}) z + 1.203$$"

# Display the equation as a mathematical expression

display(Latex(equation_latex))
[a0, a1, a2]
```

Massa:

```python
# Constants
r = 6377e3  # Earth's radius in meters
pi = np.pi

# Density function, rho(z), based on the polynomial we found earlier
def rho(z):
    return a0 + a1*z + a2*z**2

# The integrand function: rho(z) * 4pi(r + z)^2
def integrand(z):
    return rho(z) * 4 * pi * (r + z)**2

# Perform the integral from z=0 to z=25 km (converted to meters)
m, _ = quad(integrand, 0, 25e3)

# Convert m to scientific notation and format as a string
m_sci = "{:.5e}".format(m)

# Split the scientific notation string into base and exponent for LaTeX formatting
base, exponent = m_sci.split("e")

# Convert exponent from string to integer to remove any leading zeros
exponent = int(exponent)

# Create the LaTeX string using the formatted components
latex_string2 = r"$$m = {:.5f} \times 10^{{{}}} \, \text{{kg/m}}^3$$".format(float(base), exponent)

# Display the LaTeX string
display(Latex(latex_string2))
```
